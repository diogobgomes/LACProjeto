```python
# Load the spectral lines data
PATH= "C:\\Users\\jonny\\Uni\\2S\\LAC\\Projeto\\Python\\DadosRiscas_NIST\\"
Data_Hg= np.transpose( np.loadtxt(PATH+"Lines_Hg.csv", delimiter=";") )
Data_Kr= np.transpose( np.loadtxt(PATH+"Lines_Kr.csv", delimiter=";") )
Data_Ne= np.transpose( np.loadtxt(PATH+"Lines_Ne.csv", delimiter=";") )
Data_He= np.transpose( np.loadtxt(PATH+"Lines_He.csv", delimiter=";") )
Data_Ar= np.transpose( np.loadtxt(PATH+"Lines_Ag.csv", delimiter=";") )
# Load the experimental data
PATH="C:\\Users\\jonny\\Uni\\2S\\LAC\\Projeto\\DadosExperimentais\\PraiseThe"
Hg= np.loadtxt( PATH+"TrueMercurio" )
Ar= np.loadtxt( PATH+"Argon" )
He= np.loadtxt( PATH+"Helio" )
Ne= np.loadtxt( PATH+"Neon" )
Kr= np.loadtxt( PATH+"Kripton" )
# Creating the pixel array to then apply the fit to
pix= np.arange( len(Hg) )
# Factory given calibration
AjC1= np.array( (3.001082861E2, 2.722448916, -1.391971133E-3, -6.051824529E-6) )
AjC2= np.array( (3.053079780E2, 2.717314659, -1.299411180E-3, -6.856084818E-6) )
AjC3= np.array( (3.101094824E2, 2.707926217, -1.305309962E-3, -6.704495854E-6) )
# A ultima entrada é a mais recente
NossoV2_2= np.array((315.0146487662561,2.358142112760417,0.0021943508574767882,
                     -1.5014164242914622e-05))
NossoV2_3= np.array((304.65448808814426,+2.771446006047254,-0.002168093976511214,
                     -1.9434354047095735e-06))
NossoV2_3_5grau= np.array((315.57001167007337,+2.17551118793287,+0.008588338277613293,
                           -8.62291127124621e-05,+2.9828358440805296e-07,-3.9133329397982944e-10))
```

```python
def Versus(Coef, ExpData, LinesData):
    
    pix= np.arange( len(ExpData) )
    F= np.polynomial.polynomial.Polynomial(Coef)
    plt.figure(figsize=(16,9))
    plt.plot( F(pix), ExpData )
    plt.plot( LinesData[0], LinesData[1] ,"o" )
    plt.show()

# Esta func compara os dados utilizando o ajuste utilizando um polinomio com coeficientes Coef
#com a os pontos da base de dados do NIST de forma grafica
def AllVersus(Coef):
    
    pix= np.arange( len(He) )
    F= np.polynomial.polynomial.Polynomial(Coef)
    
    plt.figure(figsize=(16,9))
    plt.title("He")
    plt.plot( F(pix), He )
    plt.plot( Data_He[0], Data_He[1] ,"o" )
    plt.xlim(F(pix[0]),F(pix[-1]))
    plt.show()
    
    plt.figure(figsize=(16,9))
    plt.title("Ne")
    plt.plot( F(pix), Ne )
    plt.plot( Data_Ne[0], Data_Ne[1] ,"o" )
    plt.xlim(F(pix[0]),F(pix[-1]))
    plt.show()
    
    plt.figure(figsize=(16,9))
    plt.title("Kr")
    plt.plot( F(pix), Kr )
    plt.plot( Data_Kr[0], Data_Kr[1] ,"o" )
    plt.xlim(F(pix[0]),F(pix[-1]))
    plt.show()
    
    plt.figure(figsize=(16,9))
    plt.title("Ar")
    plt.plot( F(pix), Ar )
    plt.plot( Data_Ar[0], Data_Ar[1] ,"o" )
    plt.xlim(F(pix[0]),F(pix[-1]))
    plt.show()
    
    plt.figure(figsize=(16,9))
    plt.title("Hg")
    plt.plot( F(pix), Hg )
    plt.plot( Data_Hg[0], Data_Hg[1] ,"o" )
    plt.xlim(F(pix[0]),F(pix[-1]))
    plt.show()
```

```python
plt.figure(figsize=(16.18,10))

plt.plot (pix, np.polynomial.polynomial.Polynomial(AjC1)(pix), label="AjC1" )
plt.plot (pix, np.polynomial.polynomial.Polynomial(AjC2)(pix), label="AjC2" )
plt.plot (pix, np.polynomial.polynomial.Polynomial(AjC3)(pix), label="AjC3" )
plt.plot (pix, np.polynomial.polynomial.Polynomial(NossoV2_3)(pix), label="Nosso", lw=3.5 )
plt.plot (pix, np.polynomial.polynomial.Polynomial(NossoV2_3_5grau)(pix), label="Nosso", lw=3.5 )
plt.legend()
plt.show() 
```
